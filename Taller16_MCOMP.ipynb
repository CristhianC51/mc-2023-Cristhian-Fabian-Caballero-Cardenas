```python
import statistics
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [3.5, 2.5, 3, 1.5, 2, 1.3, 1, 0.3]
xy = [0, 2.5, 6, 4.5, 8, 6.5, 6, 2.1]
x2 = [0, 1, 4, 9, 16, 25, 36, 49]

promy = statistics.mean(y)
promx = statistics.mean(x)

a1 = (((len(x))*(sum(xy)))-((sum(x))*(sum(y))))/(((len(x))*(sum(x2)))-((sum(x))**2))
a0 = promy - (a1)*promx
print("a1 = ", a1)
print("a0 = ", a0)
print(" Y = ", a0, "+", a1, "X")

x = np.arange(0,10,0.1)
y = (a1)*(x)+(a0)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
